```python
my_dict = dict(id=101, name="Erika", is_active=True, age=45)
my_dict
```

```python
def validate(my_dict):
    if  isinstance(my_dict.get('id'),int) and \
        isinstance(my_dict['name'],str) and \
        isinstance(my_dict['is_active'],bool) and \
        isinstance(my_dict['age'],int):
        return True
result = validate(my_dict)
print(result)            
```

```python
record1 = dict(id = 102, name= "Marcus", active = True, age = 34)
record2 = dict(id = 103,name = "David", active = False, age = 29)
record3 = dict(id = 104, name = "Anna", active = True, age = 41.5)
record4 = dict(id =  106, name = "Ingrid", active = "NOPE", age = 9)
```

```python
records = [record1,record2,record3,record4]
print(records)
```

```python
schema ={
    'id': int, 
    'name': str,
    'active' : bool,
    'age' :(int,float)
 }
```

```python
def validate_record(record, schema):
    for key, expected_type in schema.items():
        if key not in record:
            return False
        if not isinstance(record[key],expected_type):
            return False
    return True  
validate_record  
```

```python
def validate_records(records, schema):
    for key, expected_type in schema.items():
        if not validate_record(records,schema):
            return False
        if not isinstance(records[key],expected_type):
            return False
    return True
valid = validate_records(records,schema)
print(valid)
```
