#  Plotting basics
- matplotlib
- seaborn 
- pandas plot() method


```python
import pandas as pd
 
df  = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/titanic.csv')
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
plt.hist(df["age"],bins= 50)
plt.xlabel("age")
plt.ylabel("count")
plt.title("Histogram ages Titanic dataset")
```

** OOP approch in matplotlib

```python
fig, ax = plt.subplots(1)
ax.hist(df["age"], bins= 50)
ax.set(title= " Histogram Titanic dataframe", xlabel = "age", ylabel = "count")
```

**bar chart in matplotlib**

```python
df.columns
```

```python
classes = df["class"].value_counts().reset_index()
classes
```

```python
fig,ax = plt.subplots(1)
ax.bar(x = classes["class"], height= classes["count"])
ax.set(xlabel= "class", ylabel= "count", title= "Classes onboard on Titanic")
```

```python
cohorts = ["DE23","DE24", "AI23", "JAVA24"]
number_students = [29, 38, 28, 20]

fig, ax = plt.subplots(1)
ax.bar(x = cohorts, height= number_students)
```
